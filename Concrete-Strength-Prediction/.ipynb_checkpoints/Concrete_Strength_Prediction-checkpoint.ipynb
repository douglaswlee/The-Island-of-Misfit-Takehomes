### Decision Tree Regression

**Then, create a single decision tree model that can reasonably predict CS using only the top two most important input measurement features, as identified in the previous stage. Generate a visual representation of your decision tree model and include the test set root mean square error (RMSE).**

```python
# Additional imports for this section
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import KFold, train_test_split, cross_validate
```

```python
# Split data into training/validation and holdout test sets
X = cs_df[['Cement', 'Fly ash']]
y = cs_df['CS']
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2, 
                                                    random_state=1)
```

```python
# Use K-fold cross validation across a range of max_depth for DecisionTreeRegressor
# Collect avg RMSE for each max_depth
kf = KFold(n_splits=5, shuffle=True, random_state=1)
rmses = []
for i in range(3, 11):
    dt = DecisionTreeRegressor(criterion="mse", 
                               max_depth=i, 
                               min_samples_split=5, 
                               min_samples_leaf=5, 
                               random_state=1)
    scores = cross_validate(dt, X_train, y_train, 
                            cv=kf, scoring='neg_mean_squared_error')
    rmses.append(np.mean([np.sqrt(-x) for x in scores['test_score']]))
```

```python
# Plot and inspect for best performing max_depth
plt.plot(range(3,11), rmses)
plt.xlabel('Depth')
plt.ylabel('Mean RMSE');
```

```python
# Fit best regressor on all training data and evaluate on holdout
dt_best = DecisionTreeRegressor(criterion="mse", 
                                max_depth=4, 
                                min_samples_split=5, 
                                min_samples_leaf=5, random_state=1)
dt_best.fit(X_train, y_train)
y_pred = dt_best.predict(X_test)
RMSE = np.sqrt(np.mean((y_pred-y_test)**2))
RMSE
```

```python
dt_best.score(X_train, y_train) # R^2
```

```python
dt_best.score(X_test, y_test) # R^2
```

```python
# Visualize the tree
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dt_best, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names=X.columns)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

**In plain English, explain the visual representation you generated.**

Above is a visulization of the final Decision Tree model generated. We initially check whether `Cement` $\leq$ 159.5, splitting the data into 31 observations for which this is true and 51 for which it is false. For the 82 observations we start with, this is the "best" split possible from all thresholds of either `Cement` or `Fly ash` (it has the smallest combined error when you compare each observation to the average of the group it is split into). The 31 observations satisfying where `Cement` $\leq$ 159.5 are then split according to whether `Fly ash` $\leq$ 136.5, while the 51 observations where `Cement` > 159.5 are split according to whether `Fly ash` $\leq$ 108. This process continues until no more splitting is possible (in this case we have reached a maximum of 4 levels of splits for `max_depth`).

**In plain English, explain what the test error signifies.**

**What is the ideal depth of your tree model? How did you determine the ideal tree depth?**

**What is the accuracy score of your model? In plain English, explain what the accuracy score signifies.**

We see above that the test RMSE is ~4.67, which means that the Decision Tree model's average prediction on the data that we held out is off by $\pm$ 4.67 Mpa of 28-day compressive strength. This measure is one indicator of the accuracy of the model -- another one is the $R^2$ value of ~0.686 also provided above. This $R^2$ value means that the model, using only the input variables `Cement` and `Fly ash` explains roughly 68.6% of the variation in `CS`, the 28-day compressive strength, observed in the data (or at least the test data above).

As previously mentioned, there are 4 levels of splits for the visualization of the Decision Tree above, meaning the depth or `max_depth` of the model = 4. To make this selection, for a range of possible depths from 3 to 10, 5-fold cross validation was performed and the average validation error across all folds was computed for each depth. This error (as RMSE) was minimized when the depth (`max_depth`) was 4.

### Random Forest Regression

**Next, create a random forest model that can reasonably predict CS using any/all of the input measurement features. Generate a visual representation of your random forest model and include the test set root mean square error (RMSE).**

**How many trees does your model include? How did you determine the number of trees to include?**

**How many features does your model account for? How did you determine how many features to account for?**

```python
# Additional imports for this stage
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
# Split data now using all input features and CS
X2 = cs_df.drop(columns=['CS', 'bin'])
y2 = cs_df['CS']
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, 
                                                        test_size=0.2, 
                                                        random_state=1)
```

For this section, we will look at all the input features as possible predictors of `CS`, the 28-day compressive strength, as above. We will again split the data into a train/validation set and holdout test set. For the former, we will use 5-fold cross validation again in a grid search to find the optimal combination of number of trees (`n_estimators`) and the number of features (`max_features`), using average mean squared error across all cross validation folds for each possible combination. The possible number of features can range from 1 feature to all 7 features, while the number of trees ranges from 2 through 20.

```python
# K-fold cross validation for a grid search of trees (n_estimators) and features (max_features)
kf2 = KFold(n_splits=5, shuffle=True, random_state=1)
param_grid = {
    'n_estimators': list(range(2,21)),
    'max_features': list(range(1,8))
}

rf = RandomForestRegressor(min_samples_split=5, 
                           min_samples_leaf=5, 
                           random_state = 1, n_jobs=-1)

grid_search = GridSearchCV(estimator = rf, 
                           param_grid = param_grid, 
                           scoring='neg_mean_squared_error', 
                           cv = kf2, 
                           n_jobs = -1, 
                           verbose = 0)
```

```python
grid_search.fit(X2_train, y2_train)
```

```python
# Find out best model (number of features and trees)
best_grid = grid_search.best_params_
best_grid
```

Having completed the grid search, we see that the best Random Forest Regressor model uses 8 trees sampling up to 6 features. After fitting this best model over the training data again, we find the most important features below.

```python
# Fit best model on all training data
rf_best = RandomForestRegressor(n_estimators=8, 
                                max_features=6, 
                                min_samples_split=5, 
                                min_samples_leaf=5,
                                random_state = 1, 
                                n_jobs=-1)
rf_best.fit(X2_train, y2_train)
```

```python
# Get feature importances of relevant features and sort in ascending order
feat_imp = rf_best.feature_importances_
features = X2.columns.values
zipped = zip(features, feat_imp)
feat_imp_sort = \
    sorted(list(zipped), key=lambda tup: abs(tup[1]))[-best_grid['max_features']:]

# Plot feature importances
feat_imp_df = pd.DataFrame.from_records(feat_imp_sort)
feat_imp_df.plot.barh(figsize=(10,7))
plt.legend([])
plt.yticks(range(0, len(feat_imp_df)), feat_imp_df[0]);
```

```python
# Evaluate best model on test set
y2_pred = rf_best.predict(X2_test)
```

```python
rf_best.score(X2_train, y2_train) # Train R^2
```

```python
rf_best.score(X2_test, y2_test) # Test R^2
```

```python
np.sqrt(np.mean((y2_pred-y2_test)**2)) # Test RMSE
```

Having now evaluated the model on the holdout data, we see that its predictions are on average $\pm$ 4.27 MPa of 28-day compressive strength on this data, explaining 73.7% of its variation.
